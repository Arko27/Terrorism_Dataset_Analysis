```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
import calendar
```

#### Dataset Link: https://www.kaggle.com/datasets/itssuru/global-terrorism

```python
df = pd.read_csv('Global_Terrorism.csv')
df.head()
```

```python
df.columns.values
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
null_check = (df.isnull().sum()/df.shape[0]) * 100

null_check
```

```python
df.drop([i for i in null_check.keys() if null_check[i] > 20], axis=1, inplace=True)

df.head()
```

```python
df.columns
```

```python
print(f"Maximum Attacks in Year      : {df['iyear'].value_counts().idxmax()}")
print(f"Maximum Attacks in Month     : {calendar.month_name[df['imonth'].value_counts().idxmax()]}")
print(f"Maximum Attacks in Country   : {df['country_txt'].value_counts().idxmax()}")
print(f"Maximum Attacks in Region    : {df['region_txt'].value_counts().idxmax()}")
print(f"Maximum Attacks in City      : {df['city'].value_counts().keys()[1]}")
print(f"Success Rate of Attacks      : {(df['success'].value_counts()[1])*100//len(df['success'])} %")
print(f"Rate of Non-Suicidal Attacks : {(df['suicide'].value_counts()[0])*100//len(df['suicide'])} %")
print(f"Most Attack Type             : {df['attacktype1_txt'].value_counts().idxmax()}")
print(f"Most Target Type             : {df['targtype1_txt'].value_counts().idxmax()}")
print(f"Maximum Attacking Group      : {df['gname'].value_counts().keys()[1]}")
print(f"Maximum Attacking Weapon     : {df['weaptype1_txt'].value_counts().idxmax()}")
```

```python
x_axis = df['iyear'].value_counts().keys()
y_axis = df['iyear'].value_counts().values

plt.figure(figsize=(16,7))
plt.title('Terrorist Attacks in various Years', fontsize=20)
plt.xticks(df['iyear'].unique(), rotation=45)
plt.xlabel('Years',fontsize=15)
plt.ylabel('No. of Attacks',fontsize=15)

sns.barplot(x=x_axis,y=y_axis)
```

```python
x_axis = df['country_txt'].value_counts().keys()[:20]
y_axis = df['country_txt'].value_counts().values[:20]

plt.figure(figsize=(16,7))
plt.title('Terrorist Attacks in different Countries', fontsize=20)
plt.xticks(rotation=45)
plt.xlabel('Country',fontsize=15)
plt.ylabel('No. of Attacks',fontsize=15)

sns.barplot(x=x_axis,y=y_axis)
```

```python
plt.figure(figsize=(16,7))

plt.title('Terrorist Attacks in various Regions', fontsize=20)
sns.countplot(data=df, x=df['region_txt'])
plt.xticks(rotation=45)
plt.xlabel('Region',fontsize=15)
plt.ylabel('No. of Attacks',fontsize=15)
```

```python
x_axis = df['city'].value_counts().keys()[1:20]
y_axis = df['city'].value_counts().values[1:20]

plt.figure(figsize=(16,7))

plt.title('Terrorist Attacks in various Cities', fontsize=20)
sns.barplot(x=x_axis,y=y_axis)
plt.xticks(rotation=45)
plt.xlabel('City',fontsize=15)
plt.ylabel('No. of Attacks',fontsize=15)
```

```python
x_axis = df['weaptype1_txt'].value_counts().keys()[:5]
y_axis = df['weaptype1_txt'].value_counts().values[:5]

plt.figure(figsize=(16,7))

plt.title('Various Weapons Used in Terrorist Attacks', fontsize=20)
sns.barplot(x=x_axis,y=y_axis)
plt.xticks(rotation=45)
plt.xlabel('Weapon',fontsize=15)
plt.ylabel('No. of Attacks',fontsize=15)
```

```python
x_axis = df['gname'].value_counts().keys()[1:21]
y_axis = df['gname'].value_counts().values[1:21]

plt.figure(figsize=(16,7))

plt.title('Top 20 Terrorist groups', fontsize=20)
sns.barplot(x=x_axis,y=y_axis)
plt.xticks(rotation=90)
plt.xlabel('Group Name',fontsize=15)
plt.ylabel('No. of Attacks',fontsize=15)
```

```python
grp_names = df['gname'].value_counts().keys()
success_rate = []

for i in grp_names:
    grp_s = df.loc[df['gname'] == i]['success']
    success_rate.append(round(sum(grp_s)*100/len(grp_s),2))
```

```python
x_axis = df['gname'].value_counts().keys()[1:21]
y_axis = success_rate[1:21]

plt.figure(figsize=(16,7))

plt.title('Top 20 Terrorist groups with their Success Rate', fontsize=20)
sns.barplot(x=x_axis,y=y_axis)
plt.xticks(rotation=90)
plt.xlabel('Group Name',fontsize=15)
plt.ylabel('Success Rate (in %)',fontsize=15)
```

```python
x_axis = df['targtype1_txt'].value_counts().keys()
y_axis = df['targtype1_txt'].value_counts().values

plt.figure(figsize=(16,7))

plt.title('Different Target Types', fontsize=20)
sns.barplot(x=x_axis,y=y_axis)
plt.xticks(rotation=90)
plt.xlabel('Target',fontsize=15)
plt.ylabel('No. of Attacks',fontsize=15)
```

```python
df_india = df[df['country_txt'] == 'India']

df_india.head()
```

```python
plt.figure(figsize=(16,7))

sns.lineplot(data=df_india, x='iyear', y = 'nkill', hue= "attacktype1_txt")

plt.title('Different Weapons used in India in various Years with Kill Count', fontsize=20)
plt.xlabel('Year',fontsize=15)
plt.ylabel('Kill Count',fontsize=15)
```

```python
x_axis = df_india['provstate'].value_counts().keys()[:15]
y_axis = [df[df['provstate'] == i]['nkill'].sum()
          for i in x_axis]

plt.figure(figsize=(16,7))
plt.xticks(rotation=45)

plt.title('No. of Deaths with respect to different States in India', fontsize=20)
sns.barplot(x = x_axis, y = y_axis)
plt.xlabel('Indian States',fontsize=15)
plt.ylabel('Kill Count',fontsize=15)
```
